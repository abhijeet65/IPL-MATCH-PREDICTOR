```python
import pandas as pd
import numpy as np
df={"HardHitter":[],"FastScorer":[],"Bat_Average":[],"Economy":[],"WicketTaker":[],"Bowl_Average":[]}
pd.DataFrame(df).to_csv("Player_Rating.csv")
bat_df=pd.read_csv('batsmen_stats.csv')
bowl_df=pd.read_csv("bowlers_stats.csv")
bat_list=[]
for ind in bat_df.index:
    bat_list.append(bat_df.iloc[ind][0])
    #print(ind)
    hard_hitter=(((bat_df["4"][ind]*4)+(bat_df["6"][ind]*6))/max(bat_df["Balls"][ind],1))
    df['HardHitter'].append(hard_hitter)
    fast_scorer=((bat_df["Runs"][ind])/max(bat_df["Balls"][ind],1))
    df["FastScorer"].append(fast_scorer)
    bat_average=(bat_df["Runs"][ind]/max(1,bat_df["Dismissals"][ind]))
    df["Bat_Average"].append( bat_average)
    df["Economy"].append(0)
    df["WicketTaker"].append(0)
    df["Bowl_Average"].append(0)
df=pd.DataFrame(df,index=bat_list)
#print(df)
for i in df:
    df[i]=df[i].astype(float)
for ind in bowl_df.index:
    if bowl_df.iloc[ind][0] in bat_list:
        economy=((bowl_df["Runs"][ind])*6.00)/(max(1,bowl_df["Balls"][ind]))
        df["Economy"][bowl_df.iloc[ind][0]]=economy
        wicket_taker=(bowl_df["Balls"][ind]/max(1,bowl_df["Wickets"][ind]))
        df["WicketTaker"][bowl_df.iloc[ind][0]]=wicket_taker
        bowl_average=(bowl_df["Runs"][ind])/(max(1,bowl_df["Wickets"][ind]))
        df["Bowl_Average"][bowl_df.iloc[ind][0]]=bowl_average
    else:
        economy=((bowl_df["Runs"][ind])*6.00)/(max(1,bowl_df["Balls"][ind]))
        wicket_taker=(bowl_df["Balls"][ind]/max(1,bowl_df["Wickets"][ind]))
        bowl_average=(bowl_df["Runs"][ind])/(max(1,bowl_df["Wickets"][ind]))
        df2={"HardHitter":[0],"FastScorer":[0],"Bat_Average":[0],"Economy":[economy],"WicketTaker":[wicket_taker],"Bowl_Average":[bowl_average]}
        df2=pd.DataFrame(df2,index=[bowl_df.iloc[ind][0]])
        df=pd.concat([df,df2])
```

```python
df.to_csv("Player_Rating.csv")
```
